```python
df = df.drop(['OSOURCE'], axis=1)
df = df.drop(['ZIP'], axis=1)
```

```python
#check the percentage of null values in each columns
null_percentage = (df.isnull().sum())/len(df) *100

null_percentage
```

```python
columns_to_drop = null_percentage[null_percentage > 85].index
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df = df_cleaned
```

```python
df['GENDER']=df['GENDER'].str.replace(' ','other')
df['GENDER']=df['GENDER'].str.replace('U','other')
df['GENDER']=df['GENDER'].str.replace('J','other')
df['GENDER']=df['GENDER'].str.replace('C','other')
df['GENDER']=df['GENDER'].str.replace('A','other')
```
